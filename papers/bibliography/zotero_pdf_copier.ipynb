```python
import pandas as pd
import os
import shutil
```

```python
# Define the source and target directories
source_directory = "c:\\Users\\Teacher\\Zotero\\storage"
target_directory = "c:\\Users\\Teacher\\Downloads\\Zotero"
```

```python
df =pd.read_csv(os.path.join(target_directory, 'Suicide_papers.csv'))
```

```python
df.head()
```

```python
file_paths = df['File Attachments'].values
```

```python
# Ensure the target directory exists
os.makedirs(target_directory, exist_ok=True)


# Function to copy files
def copy_files(file_list, dst_folder):
    for file_path in file_list:
        # Ensure the path is a string
        if not isinstance(file_path, str):
            print(f"Invalid path type: {file_path} (type {type(file_path)})")
            continue
        
        # Check if the file is a PDF and exists before copying
        if file_path.lower().endswith(".pdf"):
            if os.path.exists(file_path):
                try:
                    shutil.copy2(file_path, dst_folder)
                    print(f"Copied {file_path} to {dst_folder}")
                except Exception as e:
                    print(f"Failed to copy {file_path}. Error: {e}")
            else:
                print(f"File {file_path} does not exist.")
        else:
            print(f"Skipped non-PDF file: {file_path}")

# Copy the files
copy_files(file_paths, target_directory)

print("File copying completed.")

```
