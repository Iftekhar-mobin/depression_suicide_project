```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Creating the data frame for the given table
data = {
    "Data Point": list(range(1, 11)),
    "D+": [29.23, 73.14, 38.02, 44.59, 4.00, 65.16, 59.04, 71.18, 66.45, 43.42],
    "D-": [54.44, 62.18, 48.02, 38.47, 75.26, 12.89, 17.08, 20.56, 12.69, 40.76]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(12, 6))
sns.lineplot(x="Data Point", y="D+", data=df, marker='o', label='Ideal Solution D+')
sns.lineplot(x="Data Point", y="D-", data=df, marker='o', label='Anti Ideal Solution D-')
# plt.title('Euclidean Distance to the Ideal Solution and the Anti-Ideal Solution')
plt.xlabel('Data Point', fontsize=16)
plt.ylabel('Distance', fontsize=16)
plt.legend(title='Distance Type', fontsize=16)
plt.grid(False)

# Customize axis borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_linewidth(1)
plt.gca().spines['bottom'].set_linewidth(1)

plt.tight_layout()
plt.show()
```

```python
# Creating the data frame for the given table including TOPSIS Score
data = {
    "Data Point": list(range(1, 11)),
    "D+": [29.23, 73.14, 38.02, 44.59, 4.00, 65.16, 59.04, 71.18, 66.45, 43.42],
    "D-": [54.44, 62.18, 48.02, 38.47, 75.26, 12.89, 17.08, 20.56, 12.69, 40.76],
    "TOPSIS Score": [65.1, 49.9, 55.8, 46.3, 94.9, 16.5, 22.4, 22.3, 16.0, 48.4]
}

df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(12, 6))
sns.lineplot(x="Data Point", y="D+", data=df, marker='o', label='D+')
sns.lineplot(x="Data Point", y="D-", data=df, marker='^', label='D-')
sns.lineplot(x="Data Point", y="TOPSIS Score", data=df, marker='>', label='TOPSIS Score', linestyle='--')
plt.title('Euclidean Distance to the Ideal Solution, Anti-Ideal Solution, and TOPSIS Score')
plt.xlabel('Data Point')
plt.ylabel('Values')
plt.legend(title='Metric')
plt.grid(False)

# Customize axis borders
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_linewidth(1)
plt.gca().spines['bottom'].set_linewidth(1)

plt.tight_layout()
plt.show()
```

```python
# Creating the data frame for the given table including Data Point
data_rank = {
    "Rank": list(range(1, 11)),
    "Data Point": [5, 1, 3, 2, 10, 4, 7, 8, 6, 9],
    "TOPSIS Score": [0.949, 0.651, 0.558, 0.499, 0.484, 0.463, 0.224, 0.223, 0.165, 0.160]
}

df_rank = pd.DataFrame(data_rank)

# Plotting the data
plt.figure(figsize=(14, 8))
barplot = sns.barplot(x="Rank", y="TOPSIS Score", data=df_rank, palette="viridis")

# Creating a twin Axes sharing the xaxis
ax2 = plt.gca().twinx()

# Plotting the line chart adjacent to bars
lineplot = sns.lineplot(x="Rank", y="TOPSIS Score", data=df_rank, color="red", marker='o', ax=ax2)
lineplot.set(ylim=(0, 1))  # Set the y-axis limit for the line plot to match the bar plot

plt.title('TOPSIS Ranking with Data Points', fontsize=20)
plt.xlabel('Rank', fontsize=16)
barplot.set_ylabel('TOPSIS Score', fontsize=16)
lineplot.set_ylabel('Data Point DP Score (Line)', fontsize=16)

# Annotating the bar plot with Data Points
for index, row in df_rank.iterrows():
    barplot.text(index, row['TOPSIS Score'] + 0.02, f'DP {row["Data Point"]}', color='black', ha="center", fontsize=14)

# Customize axis borders
barplot.spines['top'].set_visible(False)
barplot.spines['right'].set_visible(False)
barplot.spines['left'].set_linewidth(1)
barplot.spines['bottom'].set_linewidth(1)

plt.tight_layout()
plt.show()
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.transform import cumsum
from bokeh.io import show
from bokeh.models import ColumnDataSource
from math import pi
import pandas as pd

# Data preparation
data = {
    'Risk Level': ['Low Risk', 'Mid Risk', 'High Risk'],
    'Frequency': [406, 336, 272],
    'Percentage': [40.04, 33.14, 26.82]
}

df = pd.DataFrame(data)
df['angle'] = df['Percentage'] / 100 * 2 * pi
df['color'] = ['#718dbf', '#e84d60', '#ddb7b1']

# Create a pie chart
p = figure(plot_height=350, title="Data Distribution By Risk Level", toolbar_location=None,
           tools="hover", tooltips="@Risk Level: @Percentage%", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4, 
        start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend_field='Risk Level', source=df)

p.axis.axis_label = None
p.axis.visible = False
p.grid.grid_line_color = None

# Show the plot
output_notebook()
show(p)
```
