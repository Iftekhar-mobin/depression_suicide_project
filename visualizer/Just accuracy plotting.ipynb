```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the classifiers and their corresponding accuracies
classifiers = ["SVM", "Linear SVM", "KNN", "Naive Bayes", "Logistic Regression", "PCA", "ANN", "CNN"]
accuracies = [98.37, 89.16, 94.03, 95.75, 89.64, 89.00, 95.75, 97.26]

# Create a bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=classifiers, y=accuracies, palette="viridis")

# Add title and labels
plt.title('Classifier Accuracies')
plt.xlabel('Classifier')
plt.ylabel('Accuracy (%)')

# Set y-axis to log scale
plt.yscale('log')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
