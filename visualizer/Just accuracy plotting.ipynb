```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the classifiers and their corresponding accuracies
classifiers = ["SVM", "Linear SVM", "KNN", "Naive Bayes", "Logistic Regression", "PCA", "ANN", "CNN"]
accuracies = [98.37, 89.16, 94.03, 95.75, 89.64, 89.00, 95.75, 97.26]

# Create a bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=classifiers, y=accuracies, palette="viridis")

# Add title and labels
plt.title('Classifier Accuracies')
plt.xlabel('Classifier')
plt.ylabel('Accuracy (%)')

# Set y-axis to log scale
plt.yscale('log')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a DataFrame with all the performance metrics
data = {
    "Model": ["ResNet-101", "Inception-V3", "VGG-16", "VGG-19", "ViT", "MobileNetV2"],
    "Accuracy": [91.13, 91.40, 92.61, 90.19, 91.54, 95.10],
    "Precision": [91.54, 91.45, 92.68, 90.82, 92.06, 95.18],
    "Recall": [91.13, 91.40, 92.61, 90.19, 91.54, 95.10],
    "F1-Score": [91.17, 91.41, 92.63, 90.23, 91.57, 95.11],
    "Cohen's Kappa": [82.17, 82.57, 85.04, 80.34, 83.01, 90.08],
#     "Test Loss": [0.2106, 0.1977, 0.1905, 0.2456, 0.1994, 0.1293],
    "Test Accuracy of Normal": [94, 91, 93, 95, 95, 96],
    "Test Accuracy of Pneumonia": [88, 91, 92, 86, 88, 94],
    "Overall Test Accuracy": [91.13, 91.40, 92.61, 90.19, 91.54, 95.10]
}

df = pd.DataFrame(data)

# Melt the DataFrame to plot multiple metrics in the same chart
df_melted = pd.melt(df, id_vars=["Model"], var_name="Metric", value_name="Value")

# Creating a bar plot for each metric using seaborn
plt.figure(figsize=(12, 8))
# https://seaborn.pydata.org/tutorial/color_palettes.html
# Spectral
# rocket_r
# Set2
# Paired
sns.barplot(x="Model", y="Value", hue="Metric", data=df_melted, palette="Spectral", linewidth=0.5, edgecolor='black')

# Set y-axis to log scale
plt.yscale('log')

plt.title("Performance Metrics of Deep Learning Models")
plt.xlabel("Model")
plt.ylabel("Value")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```
